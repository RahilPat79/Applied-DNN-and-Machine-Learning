```python
import os
import pandas as pd

path = "./data/"
filename_read = os.path.join(path,'auto-mpg.csv')
filename_write= os.path.join(path,'mod2-2output.csv')
df= pd.read_csv(filename_read,na_values=['NA','?'])
#print(df[0:5])
```

```python
df = df.select_dtypes(include=['int','float'])
print(df[0:5])
```

```python
headers = list(df.columns.values)
print(headers)
```

```python
fields=[]
for field in headers:
    fields.append({
        'name':field,
        'mean':df[field].mean(),
        'variance':df[field].var(),
        'std deviation':df[field].std()
    })

print(fields)  
```

```python
import numpy as np
df= pd.read_csv(filename_read,na_values=['NA','?'])
np.random.seed(45)
df=df.reindex(np.random.permutation(df.index))
df.reset_index(inplace=True, drop=True)
df

```

```python
df = df.sort_values(by='name',ascending=True)
print('The 5th car is {}'.format(df['name'].iloc[4]))
print(df[0:7])
```

```python
df.insert(5,'weight_kg',(df['weight']*0.465321).astype(int))
df[0:6]
```

```python
from scipy.stats import zscore
df['mpg']=zscore(df['mpg'])
df[0:5]
```

```python
med =df['horsepower'].median()
df['horsepower']=df['horsepower'].fillna(med)
print('Horse has na {}'.format(pd.isnull(df['horsepower'].values.any())))
```

```python
hp = df['horsepower']
name = df['name']
result = pd.concat([name,hp,df['cylinders']],axis=1)
result = result.sort_values(by='horsepower',ascending = True)
result[0:25]
```
